```python
from transformers import pipeline
from tqdm.notebook import tqdm
```

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")#GPU
```

```python
candidate_labels = ["Color Scheme", "Uniformity","Loading Speed","Content","White Space", "Interactivity", "Screen Size", "Readability","Accessibility","Navigation"]
```

```python
import pandas as pd
```

```python
# File Directory for both the train and test
df =pd.read_csv("gdrive/MyDrive/dataset_interface/mydataset_Interface.txt")
df.head()
```

```python
df1=df.iloc[:, 0]
df1
```

```python
from re import search

for ind in df1:
    if search('F:', ind) :
        updatedind=ind.replace('F:', '')
        #print(updatedind)
        ############################
        withoutF=classifier(updatedind, candidate_labels)
        t_label=withoutF["labels"][0]
        t_sequence=withoutF["sequence"]
        print(t_label,':',t_sequence)
```
